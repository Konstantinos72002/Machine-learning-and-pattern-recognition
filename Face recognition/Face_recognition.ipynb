```python
import math

def f1(x,w):
  return 1/(1+math.e**(-(w[0] + w[1]*x)))

def f1_partial_0(x,w):
  return (math.e**(-(w[0] + w[1]*x)))/(1 + math.e**(-(w[0] + w[1] * x)))**2

def f1_partial_1(x,w):
  return (x * math.e**(-(w[0] + w[1]*x)))/(1 + math.e**(-(w[0] + w[1] * x)))**2

def L1(x,y,w):
  return (-1/x.shape[0]) * sum([y[i] * math.log2(f1(x[i],w)) + (1 - y[i]) * math.log2(1 - f1(x[i],w)) for i in range(0, x.shape[0])])

def L1_partial_0(x,y,w):
  return -(1/x.shape[0]) * sum([(y[i] * (1/math.log(2)*f1(x[i],w)) * f1_partial_0(x[i],w) - (1 - y[i]) * 1/(math.log(2) * (1 - f1(x[i],w))) * f1_partial_0(x[i],w)) for i in range(0, x.shape[0])])

def L1_partial_1(x,y,w):
  return -(1/x.shape[0]) * sum([(y[i] * (1/math.log(2)*f1(x[i],w)) * f1_partial_1(x[i],w) - (1 - y[i]) * 1/(math.log(2) * (1 - f1(x[i],w))) * f1_partial_1(x[i],w)) for i in range(0, x.shape[0])])

def sigma(x):
  if x > 0:
    return x
  else:
    return 0

def sigma_partial(x):
  if x > 0:
    return 1
  else:
    return 0

def f2(x,w):
  return w[0] + w[1]*sigma(w[2] + w[3] * x)

def f2_(x,w):
  return w[0] + w[1]*(1)

def f2_partial0(x,w):
  return 1

def f2_partial1(x,w):
  return sigma(w[2] + w[3] * x)

def f2_partial2(x,w):
  return w[1] * sigma_partial(w[2] + w[3] * x)

def f2_partial3(x,w):
  return w[1] * x * sigma_partial(w[2] + w[3] * x)

def L2(x,y,w):
  cost = (1/x.shape[0]) * sum([(y[i] - f2(x[i],w))**2 for i in range(0, x.shape[0])])
  return cost

def L2_partial(x,y,w,partial):
  if partial == 0:
    return (1/x.shape[0]) * sum([2*(y[i] - f2(x[i],w)) * (-f2_partial0(x[i],w)) for i in range(0, x.shape[0])])
  elif partial == 1:
    return (1/x.shape[0]) * sum([2*(y[i] - f2(x[i],w)) * (-f2_partial1(x[i],w)) for i in range(0, x.shape[0])])
  elif partial == 2:
    return (1/x.shape[0]) * sum([2*(y[i] - f2(x[i],w)) * (-f2_partial2(x[i],w)) for i in range(0, x.shape[0])])
  else:
    return (1/x.shape[0]) * sum([2*(y[i] - f2(x[i],w)) * (-f2_partial3(x[i],w)) for i in range(0, x.shape[0])])


def gradient_descent(x,y,w1,w2,function,lr,iterations,sceduler):

  if function == 1:
    cost = [L1(x,y,w1)]
    for i in range(iterations):
      p0 = L1_partial_0(x,y,w1)
      p1 = L1_partial_1(x,y,w1)
      w1[0] = w1[0] - (lr * p0)
      w1[1] = w1[1] - (lr * p1)
      cost.append(L1(x,y,w1))
    return cost,w1
  if function == 2:
    cost = [L2(x,y,w2)]
    for i in range(iterations):
      if sceduler == 1:
        if i == 101:
          lr = 5 * 1e-4
        if i == 141:
          lr = 1e-4
      if sceduler == 2:
        lr = lr = 1e-6
      p0 = L2_partial(x,y,w2,0)
      p1 = L2_partial(x,y,w2,1)
      p2 = L2_partial(x,y,w2,2)
      p3 = L2_partial(x,y,w2,3)
      w2[0] = w2[0] - (lr * p0)
      w2[1] = w2[1] - (lr * p1)
      w2[2] = w2[2] - (lr * p2)
      w2[3] = w2[3] - (lr * p3)
      cost.append(L2(x,y,w2))
    return cost,w2
```

```python
import cv2
from google.colab.patches import cv2_imshow
from sklearn.preprocessing import StandardScaler
import numpy as np

def loadImages(path,set):

  map = {'Set_1' : range(1,8), 'Set_2' : range(8,20), 'Set_3' : range(20,32), 'Set_4' : range(32,46), 'Set_5' : range(46,65)}

  list_data = []
  label = []
  for i in range(1,11):
    person_number = str(i).zfill(2)
    for j in map[set]:
      total_path = path + 'person' + person_number + '_' + str(j).zfill(2) + '.png'
      im = cv2.imread(total_path, 0)
      fi = np.array(im.reshape(-1))
      list_data.append(fi)
      label.append(i)

  scaler = StandardScaler()
  scaler.fit(list_data)
  data = scaler.transform(list_data)
  return (label,data)
```

# Pca implementation

```python
import numpy as np

class PCA:
    def __init__(self, num_components):
        self.num_components = num_components
        self.eigenvectors = None
        self.mean = None

    def fit(self, X):

        self.mean = np.mean(X, axis=0)
        # Step-1: Mean centering
        X_meaned = X - self.mean

        # Step-2: Calculate covariance matrix
        cov_mat = np.cov(X_meaned, rowvar=False)

        # Step-3: Eigen decomposition
        eigen_values, eigen_vectors = np.linalg.eigh(cov_mat)

        # Step-4: Sort eigenvalues and corresponding eigenvectors
        sorted_index = np.argsort(eigen_values)[::-1]
        sorted_eigenvalues = eigen_values[sorted_index]
        sorted_eigenvectors = eigen_vectors[:, sorted_index]

        # Step-5: Select top 'num_components' eigenvectors
        self.eigenvectors = sorted_eigenvectors[:, 0:self.num_components]

    def transform(self, X):
        # Step-1: Mean centering
        X_meaned = X - self.mean

        # Step-6: Project data onto the selected eigenvectors
        X_reduced = np.dot(self.eigenvectors.T, X_meaned.T).T

        return X_reduced

    def get_eigenvectors(self):
          return self.eigenvectors

    def inverse_transform(self, X_reduced):
        # Step-7: Project reduced data back onto original dimensions
        X_original = np.dot(X_reduced, self.eigenvectors.T) + self.mean

        return X_original
```
